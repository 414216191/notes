## np.unique

找出某个属性中有唯一值的行

```python
_,index_,count_ = np.unique(data['var_0'],return_counts=True,return_index=True)
# 记录该列中的每一个值的状态
col_state = np.zeros(len(data['var_0']))
col_state[index_[count_==1]]=1
# 选出那些唯一值的行idx
np.argwhere(col_state>0)[:,0]
```

## 找出与测试样例具有相同特征值,且该特征值仅出现过一次的训练样例

```python
a = pd.DataFrame([[0,1,2],[2,3,4],[4,5,6]],columns=['a','b','c'])
b = pd.Series([0,3,5]).rename({0:'a',1:'b',2:'c'})
```

```python
t = a==b
t
```

```python
features = np.sum(t,axis=0)==1
#features = features[features==True]
features
```

```python
# 选出那些仅出现一次的样例idx
idx_mask = np.any(t.loc[:,features],axis=1)
idx_mask
a.iloc[np.argwhere(idx_mask)[:,0]]
```

## Logging包

```python
import logging

def get_logger():
    FORMAT = '[%(levelname)s]%(asctime)s:%(name)s:%(message)s'
    logging.basicConfig(format=FORMAT)
    logger = logging.getLogger('main')
    logger.setLevel(logging.DEBUG)
    return logger

logger = get_logger()
```
